```python
#Creating output of class with two classses faulty and non-faulty
df=raw_df[raw_df['X16']==0.8]['Y']

print(df.mean(),df.std(),df.max(),df.min())
updt_df=[]
def converter_to_class(df):
    mean = df.mean()
    std = df.std()
    for value in df:
        if (value<=(mean+2*std) and value>=(mean-2*std)):
            updt_df.append(1)
        else:
            updt_df.append(0)
    return updt_df
converter_to_class(df)
plt.figure(figsize=(10, 10)) 
plt.subplot(1,2,1)
plt.plot(np.arange(3150),df)
plt.subplot(1,2,2)
plt.hist(updt_df,bins=[-0.5, 0.5, 1.5], rwidth=0.8, align='mid')
plt.xticks([0, 1])
```
